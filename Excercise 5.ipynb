```python
chipo['Price'] = chipo['Price'].str.replace('$', '').astype(float)
```

```python
chipo['total'] = chipo['Game Length (Hours)'] * chipo['Price']
chipo.head(20)
```

```python
def categorize_price(Price):
    if Price < 5:
        return 'Low'
    elif Price < 10:
        return 'Medium'
    else:
        return 'High'
# We can use the .apply() method to create a new column 'price_group' based on the 'item_price' column.
chipo['price_group'] = chipo['Price'].apply(categorize_price)
chipo.head()
```
